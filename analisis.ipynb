```python
def clearColumns(df):
    patterns = ['_url', '_at', 'has_']
    for col in df.columns:
        for pattern in patterns:
            if pattern in col:
                df.drop(col, axis=1, inplace=True)

```

```python
## sort by stargazers_count
df = pd.read_json('json/language/all.json')


df.sort_values(by='stargazers_count', ascending=False, inplace=True)

## filter by open_issues > 1000

df = df[df['open_issues'] > 1000]

df.head()
```

```python
def developers_to_db():
    cursor = cnx.cursor()
    developers = []
    rows = df.shape[0]
    for i in range(rows):
        row = df.iloc[i]
        developer = row['owner']
        developers.append(developer)

        query = "INSERT INTO developers (id, login, html_url, type) VALUES (%s,%s,%s,%s)"
        values = (developer['id'], developer['login'], developer['html_url'], developer['type'])
        try :
            cursor.execute(query, values)
            cnx.commit()

        except mysql.connector.Error as err:
            print(err)
            pass
    cursor.close()
```

```python
def get_important_fields(row) -> list:
    important_fields = []
    for column in important_columns:
        if column in list(row.keys()):
            if (column == 'owner'):
                important_fields.append(row[column]['id'])
            else:
                important_fields.append(row[column])
    return important_fields
def repositories_to_db():
    cursor = cnx.cursor()
    rows = df.shape[0]
    for i in range(rows):
        row = df.iloc[i].to_dict()
        important_fields = get_important_fields(row)

        query = "INSERT INTO repositories (id, full_name, owner_id, size, watchers_count, stargazers_count, description, url, name, language, forks_count, open_issues_count, created_at, pushed_at) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"
        try :
            cursor.execute(query, important_fields)
            cnx.commit()

        except mysql.connector.Error as err:
            print(err)
            pass
    cursor.close()
    
```

## All issues (Issues + Pulls)

```python
issuesAndpulls = pd.read_json('json/fastapi/issues/all.json')
issuesAndpulls.columns
```

```python
# get pull_requests 
allIssues = issuesAndpulls[issuesAndpulls['pull_request'].isna()]
openIssues = allIssues[allIssues['state'] == 'open']
closedIssues = allIssues[allIssues['state'] == 'closed']
print({
    'all': len(allIssues),
    'open': len(openIssues),
    'closed': len(closedIssues)
})
```

## Get all pull request from API


```python
pulls = pd.read_json('json/fastapi/pulls/all.json')
pulls.columns
```

```python
mergedPulls = pulls[pulls['merged_at'].notna()]
openPulls = pulls[pulls['merged_at'].isna() & pulls['closed_at'].isna()]
closedPulls = pulls[pulls['closed_at'].notna()]
print({
    'all': len(pulls),
    'merged': len(mergedPulls),
    'open': len(openPulls),
    'closed': len(closedPulls),
})
```
